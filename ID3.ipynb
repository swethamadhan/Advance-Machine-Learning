```python
from chefboost import Chefboost as chef
```

```python
import pandas as pd

df = pd.read_csv("diabetess.csv")
```

```python
df
```

```python
config = {'algorithm': 'ID3'}
model = chef.fit(df, config = config, target_label = 'Outcome')
```

```python
df
```

```python
prediction = chef.predict(model, param = [6, 148, 72, 35,0,33.6,0.627,50])
```
