```python
from chefboost import Chefboost as chef
```

```python
import pandas as pd

df = pd.read_csv("2015.csv")
```

```python
df
```

```python
config = {'algorithm': 'CART'}
model = chef.fit(df, config = config, target_label = 'Dystopia Residual')
```

```python
df
```

```python
prediction = chef.predict(model, ['Iceland','Western Europe',2,7.561,0.04884,1.30232,1.40223,0.94784,0.62877,0.14145,0.43630,2.70201])
```
